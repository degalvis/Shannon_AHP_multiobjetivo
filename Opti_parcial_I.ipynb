```python
#METHOD PROGRAMED SHANNON
import numpy as np
import pandas as pd
import math

# Leer el archivo CSV dentro de un dataframe de pandas
df = pd.read_csv('/content/datos_proyectos.csv', delimiter = ';', encoding='latin1')

# Convertir el dataframe a un array de numpy------------------------------------------------
df = df.drop(df.columns[0], axis=1)
desicionMatrix = df.values
n, m = desicionMatrix.shape

#Vector de las sumas de las columnas (criterios)--------------------------------
sumas = []
for j in range(m):
  sumas.append(sum(desicionMatrix[:,j]))

#Normalizando la matriz inicial ------------------------------------------------
normalizada=np.ndarray((n,m))
j = 0;
while j < len(sumas):
  i = 0;
  while i < len(desicionMatrix):
    normalizada [i][j] = desicionMatrix[i][j] / sumas[j]
    i += 1
  j += 1

#Calcular la entropía---------------------------------------------------------------
#Suma de matriz normalizada por el logaritmo
entropia = np.ndarray((n, m))
i = 0
while i < n:
  j = 0
  while j < m:
    if normalizada[i][j] == 0:
      log = 0
    else:
      log = math.log(normalizada[i][j])
    entropia[i][j] = normalizada[i][j] * log
    j += 1
  i += 1

#Calcular valor de la entropía
valorEntropia = []
k = 1/math.log(m)

for j in range(m):

  valorEntropia.append(sum(entropia[:,j]) * -k)

#Pesos por alternativas --------------------------------------------------------
#1-e_j
uno_ej = []

for i in valorEntropia:
  uno_ej.append(1 - i)

#sumas 1-e_j
sumas_uno_ej = 0

for i in uno_ej:
  sumas_uno_ej += i
w_i = []
for i in uno_ej:
  temporal = i / sumas_uno_ej
  w_i.append(temporal)

pesos = np.zeros(m)
i = 0
while i < n:
  j = 0
  temp = 0
  while j < m:
    temp += w_i[j] * normalizada[i][j]
    j += 1
  pesos[i] = temp
  i += 1

print('pesos', pesos)
```
